```python
import numpy as np
import matplotlib.pyplot as plt
Z = 18
kT = 0.616879
```

```python
def load_file(fname):
    f = open(fname)
    fl = f.readlines()
    f.close()
    
    clean = []
    for line in fl:
        if "@" not in line and "#" not in line:
            clean.append(line.split())
    clean = np.array(clean, dtype='float')
    return clean
```

# Nanoparticle

```python
np_bar = load_file("NP/T1-N0-I0_bar.xvg")
fig = plt.figure()
ax = plt.axes()
ax.set_xlim((0,31))
ax.tick_params(labelsize=Z)
ax.set_xlabel("$\lambda_i$", fontsize=Z)
ax.set_ylabel("Step free energy\n($kCal/mol$)", fontsize=Z)
ax.errorbar(np_bar[:,0], np_bar[:,1]*kT, yerr=np_bar[:,2], lw=2, c='r')
plt.show()
```

```python
np_barint = load_file("NP/T1-N0-I0_barint.xvg")
fig = plt.figure()
ax = plt.axes()
ax.set_xlim((0,31))
ax.tick_params(labelsize=Z)
ax.set_xlabel(r"$\lambda_i$", fontsize=Z)
ax.set_ylabel("Cumulative free\nenergy ($kCal/mol$)", fontsize=Z)
ax.errorbar(np_barint[:,0], np_barint[:,1]*kT, lw=2, c='b')
plt.show()
np_dg = np_barint[-1,1]*kT
print("Solvation energy: {:.2f} kcal/mol".format(np_dg))
```

```python
np_hist = load_file("NP/T1-N0-I0_hist.xvg")
np_hist = np.reshape(np_hist, (len(np_hist)//200, 200, 2))

ndx_dh = np.arange(-1, 125, 4)
ndx_dh[0] = 0
ndx_dhdl = []
for i in range(126):
    if i not in ndx_dh:
        ndx_dhdl.append(i)

dh = np_hist[ndx_dh]
dhdl = np_hist[ndx_dhdl]

subplot_kw = {}
fig, axs = plt.subplots(figsize=(14,5), ncols=1, nrows=2, sharey=True, gridspec_kw={'wspace':0.3})
for ax in axs:
    ax.tick_params(labelsize=Z)
axs[0].set_xlabel(r"$\Delta H$", fontsize=Z)
axs[0].set_ylabel("Counts", fontsize=Z)
axs[1].set_xlabel("$\Delta H$/$\Delta \lambda$", fontsize=Z)
axs[1].set_ylabel("Counts", fontsize=Z)
axs[1].set_ylim((0,6000))
for hist in dh:
    axs[0].errorbar(hist[:,0], hist[:,1], lw=2)
for hist in dhdl:
    axs[1].errorbar(hist[:,0], hist[:,1], lw=2)
plt.show()
```
